```python
import math
import numpy as np
from sklearn import datasets, linear_model

with open('vidu3_lin_reg.txt') as f:
    lines = f.readlines()

x_data = []
y_data = []
lines.pop(0)
num_train_data = int(len(lines)*8/10)

def get_data(start, end, lines):
    for i in range(start,end):
        splitted = lines[i].replace('\n', '').split(' ')
        splitted.pop(0)
        splitted = list(map(float, splitted))
        y_data.append([splitted[5]])
        x_data.append(splitted[0:5])
    return x_data, y_data

x_train_data, y_train_data = get_data(0,num_train_data, lines)
x_train_data = np.asarray(x_train_data)
y_train_data = np.asarray(y_train_data)
x_test_data, y_test_data = get_data(num_train_data, len(lines), lines)
x_test_data = np.asarray(x_test_data)
y_test_data = np.asarray(y_test_data)

regr = linear_model.LinearRegression(fit_intercept=False)

regr.fit(x_train_data, y_train_data)
y_lin_reg = []

for i in range(len(x_test_data)):
    y_lin_reg.append(regr.predict([x_test_data[i]])[0])
y_lin_reg = np.asarray(y_lin_reg)
result = []

for i in range(len(x_test_data)):
    result.append([y_test_data[i][0], y_lin_reg[i][0]])
result = np.asarray(result)
print(result)
```
